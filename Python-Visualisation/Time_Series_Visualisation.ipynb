```python
#Visualisation packages
import matplotlib.pyplot as plt

#Computational packages
import numpy as np
import pandas as pd

#Hide Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Companies stock data
column_names = ['Date', 'Value']
aapl = pd.read_csv('data/aapl.csv', names=column_names)
ibm = pd.read_csv('data/ibm.csv', names=column_names)
csco = pd.read_csv('data/csco.csv', names=column_names)
msft = pd.read_csv('data/msft.csv', names=column_names)

aapl_series = aapl.set_index(pd.DatetimeIndex(aapl['Date'])).iloc[:,:]
ibm_series = ibm.set_index(pd.DatetimeIndex(ibm['Date'])).iloc[:,:]
csco_series = csco.set_index(pd.DatetimeIndex(csco['Date'])).iloc[:,:]
msft_series = msft.set_index(pd.DatetimeIndex(msft['Date'])).iloc[:,:]
```

```python
# Time Series stocks on the same axes
plt.plot(aapl_series.resample("1A", how=sum), color='b', label='Apple')
plt.plot(ibm_series.resample("1A", how=sum), color='g', label='IBM')
plt.plot(csco_series.resample("1A", how=sum), color='r', label='Cisco')
plt.plot(msft_series.resample("1A", how=sum), color='m', label='Microsoft')

plt.legend(loc='upper right')
plt.xticks(rotation=60) #Orientation of the xticks
plt.rcParams["figure.figsize"] = (8,6)
plt.show()
```

```python
# Apple stock prices over full 11-year range and over shorter 2-year range

plt.subplot(2,1,1)
plt.xticks(rotation=45)
plt.title('AAPL: 2001 to 2011')
plt.plot(aapl_series.resample("1A", how=sum), color='blue')

two_year_view = aapl_series['2007':'2008']
plt.subplot(2,1,2)
plt.xticks(rotation=45)
plt.title('AAPL: 2007 to 2008')
plt.plot(two_year_view.resample("1M", how=sum), color='green')

plt.tight_layout()
plt.show()
```

```python
# Comparison between all companies Stock prices between Nov 2007 and April 2008 and Specifically for January 2008
nov_apr_appl = aapl_series['2007-11':'2008-04']
nov_apr_csco = csco_series['2007-11':'2008-04']
nov_apr_ibm = ibm_series['2007-11':'2008-04']
nov_apr_msft = msft_series['2007-11':'2008-04']

plt.subplot(2,1,1)
plt.plot(nov_apr_appl.resample("1M", how=sum), color='red', label='Apple')
plt.plot(nov_apr_csco.resample("1M", how=sum), color='blue', label='Cisco')
plt.plot(nov_apr_ibm.resample("1M", how=sum), color='green', label='IBM')
plt.plot(nov_apr_msft.resample("1M", how=sum), color='orange', label='Microsoft')
plt.title('AAPL: Nov. 2007 to Apr. 2008')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(loc = 'upper right')

jan_2008_appl = aapl_series['2008-01']
jan_2008_csco = csco_series['2008-01']
jan_2008_ibm = ibm_series['2008-01']
jan_2008_msft = msft_series['2008-01']

plt.subplot(2,1,2)
plt.plot(jan_2008_appl.resample("1B", how=sum), color='red', label='Apple')
plt.plot(jan_2008_csco.resample("1B", how=sum), color='blue', label='Cisco')
plt.plot(jan_2008_ibm.resample("1B", how=sum), color='green', label='IBM')
plt.plot(jan_2008_msft.resample("1B", how=sum), color='orange', label='Microsoft')
plt.title('AAPL: Jan. 2008')
plt.xticks(rotation=45)
plt.rcParams["figure.figsize"] = (15,10)
plt.legend(loc = 'upper right')

#Inserting Weekly level aggregation inset view
plt.axes((0.1, 0.3, 0.2, 0.1))
plt.plot(jan_2008_appl.resample("1W", how=sum), color='red', label='Apple')
plt.plot(jan_2008_csco.resample("1W", how=sum), color='blue', label='Cisco')
plt.plot(jan_2008_ibm.resample("1W", how=sum), color='green', label='IBM')
plt.plot(jan_2008_msft.resample("1W", how=sum), color='orange', label='Microsoft')
plt.xticks(rotation=45)
plt.title('Weekly Aggregation')

#Display
plt.tight_layout()
plt.show()
```

```python
# Moving averages over 30, 75, 125, and 250 days for Apple stocks

mean_30_aapl = aapl_series.resample("30D", how=sum)
mean_75_aapl = aapl_series.resample("75D", how=sum)
mean_125_aapl = aapl_series.resample("125D", how=sum)
mean_250_aapl = aapl_series.resample("250D", how=sum)

# Plot the 30-day moving average in the top left subplot in green
plt.subplot(2,2,1)
plt.plot(mean_30_aapl, 'green')
#plt.plot(aapl_series, 'k-.')
plt.xticks(rotation=60)
plt.title('30 D averages')

# Plot the 75-day moving average in the top right subplot in red
plt.subplot(2,2,2)
plt.plot(mean_75_aapl, 'red')
#plt.plot(aapl_series, 'k-.')
plt.xticks(rotation=60)
plt.title('75 D averages')

# Plot the 125-day moving average in the bottom left subplot in magenta
plt.subplot(2, 2, 3)
plt.plot(mean_125_aapl, 'magenta')
#plt.plot(aapl_series, 'k-.')
plt.xticks(rotation=60)
plt.title('125 D averages')

# Plot the 250-day moving average in the bottom right subplot in cyan
plt.subplot(2,2,4)
plt.plot(mean_250_aapl, 'cyan')
#plt.plot(aapl_series, 'k-.')
plt.xticks(rotation=60)
plt.title('250 D averages')

# Display the plot
plt.show()
```
