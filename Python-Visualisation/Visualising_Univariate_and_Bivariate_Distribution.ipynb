```python
#Computational Packages
import numpy as np
import pandas as pd

#Visualisation Packages
import matplotlib.pyplot as plt
import seaborn as sns

#Hide Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading MPG Auto Dataset
auto_dataset = pd.read_csv('data/mpg_auto_datasets.csv')
auto_dataset.info()
```

# Univariate Strip plot of 'hp' grouped by 'cyl'

```python
plt.subplot(2,1,1)
sns.stripplot(x='cyl', y='hp', data=auto_dataset)

plt.subplot(2,1,2)
sns.stripplot(x='cyl', y='hp', data=auto_dataset, jitter=True, size=3) #Using jitter and smaller point size

#Display
plt.rcParams["figure.figsize"] = (15,7)
plt.show()
```

# Univariate Swarm Plot of 'hp' grouped by 'cyl'

```python
plt.subplot(2,1,1)
sns.swarmplot(x='cyl', y='hp', data=auto_dataset)

plt.subplot(2,1,2)
sns.swarmplot(x='hp', y='cyl', data=auto_dataset, hue='origin', orient='h') #Swarm plot

# Display
plt.rcParams["figure.figsize"] = (15,7)
plt.show()
```

# Univariate Violin Plot

```python
plt.subplot(2,1,1)
sns.violinplot(x='cyl', y='hp', data=auto_dataset)

plt.subplot(2,1,2)
sns.violinplot(x='cyl', y='hp', data=auto_dataset, inner=None, color='lightgray') #Added color of lightgray

sns.stripplot(x='cyl', y='hp', data=auto_dataset, jitter=True, size=1.5) #Overlaying strip plot on the violin plot

# Display
plt.rcParams["figure.figsize"] = (15,7)
plt.show()
```

# Bivariate joint plot of 'hp' & 'mpg'

```python
sns.jointplot(x='hp', y='mpg', data=auto_dataset, color='red')

# Display
#plt.rcParams["figure.figsize"] = (18, 10)
plt.show()
```

# Different types of joint variation for bivariate analysis

```python
#Hexbin plot of the data points
sns.jointplot(x='hp', y='mpg', data=auto_dataset, kind='hex', color='blue')
plt.show()
```

```python
#Scatter plot of the data points
plt.clf()
sns.jointplot(x='hp', y='mpg', data=auto_dataset, kind='scatter', color='red')
plt.show()
```

```python
#Regression plot of the data points
plt.clf()
sns.jointplot(x='hp', y='mpg', data=auto_dataset, kind='reg', color='green')
plt.show()
```

```python
#Residual plot of the data points
plt.clf()
sns.jointplot(x='hp', y='mpg', data=auto_dataset, kind='resid', color='magenta')
plt.show()
```

```python
#Kernel density estimate plot of the data points
plt.clf()
sns.jointplot(x='hp', y='mpg', data=auto_dataset, kind='kde', color='cyan')
plt.show()
```

# Pairwise distribution plot

```python
sns.pairplot(auto_dataset)
plt.rcParams["figure.figsize"] = (15,7)
plt.show()
```

```python
#Pairwise distribution grouped by 'origin' with regression lines

sns.pairplot(auto_dataset, kind='reg', hue='origin')
plt.show()
```
