Building scalable and hyperparameter enabled pipeline to classify wine quality. The classifier used is SVM.

```python
#Loading the requisite packages
#Scikit-Learn packages
from sklearn.preprocessing import scale, StandardScaler, Imputer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#Computation and Visualization Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#Loading the requisite dataset
df = pd.read_csv('wine-quality-dataset.csv')
df.info()
```

```python
#Specifying Feature and Target Variables
y = df.quality.values
X = df.drop(['quality'], axis=1).values
```

```python
#Training and Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
# Establishing the pipeline
steps = [('scaler', StandardScaler()), ('SVM', SVC())]
pipeline = Pipeline(steps)
# Specifying the hyperparameter space
parameters = {'SVM__C':[1, 10, 100], 'SVM__gamma':[0.1, 0.01]}
```

```python
# Instantiating the GridSearchCV object
cv = GridSearchCV(pipeline, parameters, cv=5)
cv.fit(X_train, y_train)
y_pred = cv.predict(X_test)
```

```python
# Metrics Computation
print("Accuracy: {}".format(cv.score(X_test, y_test)))
print(classification_report(y_test, y_pred))
print("Tuned Model Parameters: {}".format(cv.best_params_))
```
